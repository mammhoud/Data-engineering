## Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=[10,11,12,13,14,15,16]
y=[100,170,200,270,300,370,400]
```

```python
#plot(x-axis,y-axis) function is used to draw line 
plt.plot(x,y)
plt.title('Graph One')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
A=[10,11,12,13,14,15,16]
B1=[100,170,200,270,300,370,400]
B2=[90,110,130,150,170,190,210]
```

```python
plt.plot(A,B1)
plt.plot(A,B2)
plt.title('Graph Two')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(['Data 1','Data 2'])
```

```python
#Other way to use legend
plt.plot(A,B1,label='Data 1')
plt.plot(A,B2,label='Data 2')
plt.title('Graph Two')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
```

```python
#  format 1 color,linestyle and marker are added
#marker-> specify each point with mark in the line
plt.plot(A,B1,color='blue',linestyle='-.',marker='.',label='Data 1')
plt.plot(A,B2,color='green',linestyle='--',marker='o',label='Data 2')
plt.title('Graph Two')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
```

```python
#  format 2 line width is added
plt.plot(A,B1,color='y',linestyle=':',marker='x',linewidth=2,label='Data 1')
plt.plot(A,B2,color='k',linestyle='--',marker='o',linewidth=3,label='Data 2')
plt.title('Graph Two')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid()
```

```python
#different styles available
plt.style.available
```

```python
#  format 3  grid and style are added
plt.plot(A,B1,linestyle='solid',label='Data 1')
plt.plot(A,B2,linestyle='--',label='Data 2')
plt.title('Graph Two')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True)
plt.style.use('seaborn-dark')
```

```python
#bar() function draws bar graphs vertically
#  format 1
plt.bar(A,B1)
plt.title('Graph Two (Bar Chart)')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.style.use('Solarize_Light2')
```

```python
#  format 2  width of the bar is specified
plt.bar(A,B1,width=0.35)
plt.title('Graph Two (Bar Chart)')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.style.use('Solarize_Light2')
```

```python
#  example 2 in bar()
x=["A","B","C","D"]
y=[2,6,5,1]
plt.bar(x,y,width=0.5,color='plum')
plt.title("Students Grades")
plt.xlabel("Grades")
plt.ylabel("Number of students")
plt.style.use('seaborn-dark')
plt.grid(True)
```

```python
#barh() function draws bar graphs horizontally
#instead of width here we use height
x=["A","B","C","D"]
y=[2,6,5,1]
plt.barh(x,y,height=0.5,color='cornflowerblue')
plt.title("Students Grades")
plt.xlabel("Grades")
plt.ylabel("Number of students")
plt.style.use('seaborn-dark')
plt.grid(True)
```

```python
#scatter() function draws a scatter plot.
x=[1,3,4,5,7,12,15,16,17,21]
y=np.linspace(100,150,10)
plt.scatter(x,y,color='lightgreen')
plt.title("Scatter Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
```

```python
#pie() function draws pie charts
#  format 1
p=[6,5,3,1]
l=["Cat","Dog","Bird","Fish"]
c=['deepskyblue','skyblue','royalblue','navy']
plt.pie(p,labels=l,colors=c)
```

```python
#  format 2 start angle and explode are added
p=[6,5,3,1]
l=["Cat","Dog","Bird","Fish"]
c=['deepskyblue','skyblue','royalblue','navy']
# each value in explodes represents how far from the center each part
explodes=[0,0.1,0.3,0]
plt.pie(p,labels=l,startangle=90,colors=c,explode=explodes)
```
